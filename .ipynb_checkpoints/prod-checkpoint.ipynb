```python
def tinkoff_scraper(src, region_name):
    '''Функция для выгрузки данных Tinkoff CoronaIndex по отдельному региону
    src :   ссылка на массив данных 
    region_name : название региона, которое затем используется для именования эксель-файлов'''
    response = requests.get(src)
    data = json.loads(response.content)
    
    # создаем массив с данными о динамике индекса, доле онлайн платежей, потребительской активности
    df_total = pd.DataFrame(data = data['total']['points'], columns = ['Динамика индекса', 'Доля онлайн платежей', 'Потребительская активность'])
    df_total.insert(0, column = 'period', value = pd.date_range(start='02/01/2020', periods=len(df_total), freq='D'))
    
    # массив с данными по расходам по отраслям
    categories = ['Duty free', 'Авиабилеты', 'Авто услуги', 'Аптеки', 'Аренда авто', 
                  'Гос. сборы', 'Дом, ремонт', 'Ж/д билеты', 'Животные', 'Искусство', 
                  'Кино', 'Книги', 'Красота', 'Мед. услуги', 'Музыка', 'Нко', 'Образование', 
                  'Одежда, обувь', 'Отели', 'Развлечения', 'Рестораны', 'Связь, телеком', 'Сервис. услуги', 
                  'Спорттовары', 'Сувениры', 'Супермаркеты', 'Топливо', 'Транспорт', 'Турагентства', 
                  'Фастфуд', 'Фото/видео', 'Цветы', 'Частные услуги']

    df_categories = pd.DataFrame()
    for item in categories:
        df_categories[item] = data['categories'][item]['points']
    df_categories.insert(0, column = 'period', value = pd.date_range(start='02/01/2020', periods=len(df_categories), freq='D'))    
    
    # список всех показателей оборота
    business_turnover = ['E-commerce', 'IT-услуги и разработка ПО', 'Автоуслуги', 'Аренда и лизинг', 'Бытовые услуги', 
     'Гостиницы', 'Грузоперевозки', 'ЖКХ', 'Здравоохранение и соцуслуги', 'Кафе и рестораны', 
     'Консалтинг в различных сферах', 'Красота', 'Легкая промышленность', 'Научная и техническая деятельность', 
     'Обрабатывающая промышленность', 'Образование', 'Обслуживание транспорта, cкладов', 'Общественные организации', 
     'Оптовая торговля компьютерами и ПО', 'Оптовая торговля продуктами', 'Оптовая торговля стройматериалами', 
     'Организация конференций и выставок', 'Пассажироперевозки', 'Почта и курьеры', 'Продуктовые магазины', 
     'Производство', 'Прочая оптовая торговля', 'Прочая розничная торговля', 'Развлечения, культура, спорт', 
     'Розничная торговля книгами и игрушками', 'Розничная торговля одеждой', 'Розничная торговля электроникой', 
     'Сельское хозяйство', 'Строительство', 'Телекоммуникации, связь, издательства', 'Туристические агенства', 
     'Услуги в сегменте недвижимости', 'Хозяйственная деятельность', 'Энергетика и водоснабжение']

    #массив с показателями оборота по отраслям
    df_business_turnover = pd.DataFrame()
    for item in business_turnover:
        df_business_turnover[item] = data['businessCategories'][item]['points']

    #засекаем дату начала отсчета и форматируем ее в гггг/мм/дд, добавляем стобец дат
    formatted_date = datetime.strftime(datetime.strptime(data['businessCategories']['E-commerce']['start'], '%d.%m.%Y'), '%m/%d/%Y')
    df_business_turnover.insert(0, column = 'period', value = pd.date_range(start=formatted_date, periods=len(df_business_turnover), freq='7D'))    

    # Пишем в эксель
    date = str(datetime.today().year) + '_' + str(datetime.today().month) + '_' + str(datetime.today().day)
    filepath = r'C:\\Users\\Egran\\YandexDisk\\database\\tinkoff coronaindex\\'
    filename = filepath + str(region_name) + '_' + str(date ) + '.xlsx'
    
    writer = pd.ExcelWriter(filename, engine='xlsxwriter')

    df_total.to_excel(writer, sheet_name='Сводные данные')
    df_categories.to_excel(writer, sheet_name='По тратам потребителей')
    df_business_turnover.to_excel(writer, sheet_name='По оборотам бизнеса')

    writer.save()
```
